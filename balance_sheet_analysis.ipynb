```python
import pandas as pd
import requests
import json
```

```python
bs = requests.get(f'https://financialmodelingprep.com/api/v3/financials/balance-sheet-statement/AAPL?period=quarter')
bs = bs.json()
bs = bs['financials']
```

```python
# Turn the json date into a dataframe
balance_sheet = pd.DataFrame.from_dict(bs)
```

# Transpose Data

```python
bs = balance_sheet.T
bs.columns = bs.iloc[0]
bs.head()
```

```python
bs.shape
```

```python
# Remove top row
cols = bs.columns
bs = bs.iloc[1:,]
#bs.head()
bs[cols]
```

```python
#Keep only 4 last 4 quarters(columns) and remove last 12 rows(liabilities & equites)
bs = bs.iloc[:12,:4]
cols = bs.columns
balance_sheet_assets = bs[cols].apply(pd.to_numeric, errors='coerce')
```

# Divide each balance sheet items in a quarter by the Total assets

```python
assetsQ1 = balance_sheet_assets.iloc[8,0]
assetsQ1
```

```python
assetsQ2 = balance_sheet_assets.iloc[8,1]
assetsQ2
```

```python
assetsQ3 = balance_sheet_assets.iloc[8,2]
assetsQ3
```

```python
assetsQ4 = balance_sheet_assets.iloc[8,3]
assetsQ4
```

```python
all_assets = [assetsQ1,assetsQ2,assetsQ3,assetsQ4]
all_assets
```

```python
assets = (bs[cols] / all_assets)*100
assets
```
